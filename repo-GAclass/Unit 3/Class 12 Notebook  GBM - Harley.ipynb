```python
import pandas as pd
import numpy as np
pd.options.plotting.backend = "plotly"
from sklearn.tree import DecisionTreeRegressor
df = pd.read_csv('/Users/harleyhoffmann/dat-02-22/ClassMaterial/Unit3/data/housing.csv')
```

```python
df
```

```python
y = df['PRICE']
X = df.drop('PRICE', axis=1)
```

```python
guess = y.mean()
```

```python
tree = DecisionTreeRegressor(max_depth=3)
```

```python
gradient = y - guess
#simple guess, calculating error - price minus average price
```

```python
gradient
```

```python
tree.get_params()
```

```python
tree.fit(X, gradient)
#we're fitting the model on the graident because we're trying to predict on the error
```

```python
tree.predict(X)
#trying to predict the error from the average
```

```python
guess += tree.predict(X)
#we're taking our guess and then adding in the predicted error
```

```python
guess
```

```python
y - guess
#numbers are now significantly smaller because we added in the expected loss to guess above
#we want to get these numbers as low as possible
```

```python
gradient = y - guess
#this is a new gradient-1
```

```python
tree.fit(X, gradient)
#fitting it on the new gradient, resulting tree little different-2
```

```python
guess += tree.predict(X)
#add in the error once more-3
```

```python
guess[:5]
```

```python
y[:5]
```

```python
#final thing
learning_rate = 0.1
```

```python
gradient = y - guess
```

```python
tree.fit(X, gradient)
```

```python
tree.predict(X)*learning_rate
#we're now updating it with the values of the predictions, scaled by the learning rate - 1/10 of the error
```

```python
guess += tree.predict(X)*learning_rate
#slowly add in our errors so that the steps we take after each round should be smaller -- like how a tree grows branches
```
